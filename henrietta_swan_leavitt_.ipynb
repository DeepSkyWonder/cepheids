# **Henrietta Swan Leavitt - Period Luminosity Relation**

**Import Python Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Download CSV file of Henrietta's Stars from GitHub**

```python
url = 'https://raw.githubusercontent.com/DeepSkyWonder/cepheids/main/heriettas-stars.csv'
```

```python
henriettas_stars_dataframe = pd.read_csv(url)
henriettas_stars_dataframe
```

**Calculate the Log of the Periods**

```python
log_periods = np.log10(henriettas_stars_dataframe['period'])
```

**Calculate a Linear Fit to minimum and maximum magnitude curvers**

```python
# Extract Min and Max Magnitudes from the Data Table
mag_max = henriettas_stars_dataframe['mag_max']
mag_min = henriettas_stars_dataframe['mag_min']

# Create Linear Models and Predictors for the Magnitude Data
model_mag_max = np.polyfit(log_periods, mag_max, 1)
predict_mag_max = np.poly1d(model_mag_max)

model_mag_min = np.polyfit(log_periods, mag_min, 1)
predict_mag_min = np.poly1d(model_mag_min)

# Create Array of Log-Period Values
log_period_range = np.arange(0, 2.3, 0.1)

# Use the Linear Model to Predict Minimum and Maximum Magnitude Values
prediction_mag_max = predict_mag_max(log_period_range)
prediction_mag_min = predict_mag_min(log_period_range)
```

**Plot the Magnitude of Henrietta's Cepheids against the Log of their Periods and display the Linear Model Fit**

```python
plt.figure(figsize=(5.2,5.2), dpi=200)
plt.plot(log_periods, mag_max, 'bo-', linewidth=1, markersize=3)
plt.plot(log_period_range, prediction_mag_max, 'r', linewidth=1)
plt.plot(log_periods, mag_min, 'bo-', linewidth=1, markersize=3)
plt.plot(log_period_range, prediction_mag_min, 'r', linewidth=1)
plt.xlabel('log period (days)')
plt.ylabel('magnitude')
plt.xticks(np.arange(0, 2.21, 0.2))
plt.yticks(np.arange(11, 16.51, 0.5))
plt.xlim([0.0, 2.2])
plt.ylim([11.0, 16.5])
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```
